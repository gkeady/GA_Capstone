#### To finish our setup let's remove the columns which our data cannot currently take advantage of as well as create any masks that we may find useful for modelling

```python
#unfortunately as we are not presently able to access better target variable data, these columns will not help us in our 
#modelling

ygo.drop(['release_date','set_codes'],axis=1, inplace=True)
```

```python
#we also appear to have had some cards slip through the net that have no listed price so we'll remove those now

ygo = ygo[ygo.lowest_price!=0]

#NB because I only noticed this at this stage, the data here has one extra column (extra secret rare) which has only zero
#values, the dataflow in the overview document does not make this mistake and so does not have the extra secret rare colummn
```

```python
#we have our price mask from the EDA and also one for excluding any cards which do not have an effect

price_mask = ygo.lowest_price <= 20

effect_mask = ygo.effect_or_flavour_text == 1
```

```python
ygo[price_mask & effect_mask].describe()
```

```python
#we can also initialise a dataframe with our masks active

ygo_mask = ygo[price_mask & effect_mask].copy()
ygo_mask.reset_index(drop=True,inplace=True)
```

### For our NLP data it makes most sense to start with a count vectoriser as the rarer terms are not necessarily more useful being boosted

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
stop.append('per')
stop.append('card')
```

```python
stop

#maybe remove itself, also, then, when, if and add effect, turn?
#overall this list is something that you could go back and forth on endlessly
```

```python
cvec_name = CountVectorizer(stop_words=stop,ngram_range=(1,3),token_pattern=r"\w[\w'\"()\@\-/]+",min_df=5,max_features=2500)
cvec_text = CountVectorizer(stop_words=stop,ngram_range=(1,2),token_pattern=r"\w[\w'()\@\-/]+",min_df=10,max_features=10000)

cvec_name.fit(ygo_mask['name'])
cvec_text.fit(ygo_mask['card_text'])
```

```python
c_name = cvec_name.transform(ygo_mask['name'])
```

```python
top_name = c_name.sum(axis=0)
pd.DataFrame(top_name, columns=cvec_name.get_feature_names()).transpose().sort_values(by=0, ascending=False)[:]
```

```python
count=0
for i in ygo.name:
    if '/' in i:
        count+=1
        print(i)
print(count)

#regex problems can be demonstrated by trying to list all D/D/D, @ignister, /Assault, or cards with "" in their name
#thus words with / in them should become single words also card names with apostrophes cause issues e.g. magician's !=
#magicians != magician, essentially compromises must be made
#in the long term potential tweaks to the vectoriser might be possible that remove a few of these issues such as using 
#nltk's stemmers https://www.elastic.co/guide/en/elasticsearch/guide/current/choosing-a-stemmer.html
```

```python
c_text = cvec_text.transform(ygo_mask['card_text'])
```

```python
top_text = c_text.sum(axis=0)
pd.DataFrame(top_text, columns=cvec_text.get_feature_names()).transpose().sort_values(by=0, ascending=False)[:]
```

```python
dfc_name = pd.DataFrame(c_name.todense(),columns=cvec_name.get_feature_names())
dfc_text = pd.DataFrame(c_text.todense(),columns=cvec_text.get_feature_names())
```

```python
X = pd.concat([ygo_mask,dfc_name,dfc_text],axis=1)
```

```python
X.info()
```

```python
X.drop(labels=['name','card_text','effect_or_flavour_text'],axis=1,inplace=True)
```

```python
y = X.pop('lowest_price')
```

## Trying Multiple Models

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.dummy import DummyRegressor
from sklearn.tree import DecisionTreeRegressor,ExtraTreeRegressor
from sklearn.svm import LinearSVR,SVR,NuSVR
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet,Lars,LassoLars,OrthogonalMatchingPursuit,PoissonRegressor,TweedieRegressor,GammaRegressor,SGDRegressor,PassiveAggressiveRegressor
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor,BaggingRegressor,ExtraTreesRegressor,RandomForestRegressor,StackingRegressor,VotingRegressor
```

```python
#train test split
X_Train,X_Test,y_Train,y_Test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=3)
```

```python
#scaler this might also be tried with a minmax scaler due to the number of categorical variables that have been dummified to
#1 or 0
scaler = StandardScaler()
X_TrSc = scaler.fit_transform(X_Train)
X_TeSc = scaler.transform(X_Test)
```

```python
models = [
    DummyRegressor(strategy='mean'),
    LinearRegression(),
    DecisionTreeRegressor(max_depth=7),
    ExtraTreeRegressor(max_depth=7),
    Ridge(alpha=1.5,solver='sag'),
    Lasso(),
    SVR(cache_size=500),
    SVR(kernel='poly',degree=2,cache_size=500),
    NuSVR(cache_size=500),
    NuSVR(kernel='poly',degree=2,cache_size=500),
    ElasticNet(),
    Lars(),
    LassoLars(),
    SGDRegressor(loss='huber',penalty='elasticnet',early_stopping=True,n_iter_no_change=10),
    PassiveAggressiveRegressor(early_stopping=True,n_iter_no_change=10),
    OrthogonalMatchingPursuit(),
    PoissonRegressor(),
    TweedieRegressor(power=1.5,link='log'),
    GammaRegressor()
]

#the dummy regressor essentially functions as the baseline prediction
#NB that the poisson, tweedie, gamma models have a different scoring metric for the percentage of deviance explained
```

```python
for model in models:
    print(model)
    print()
    model.fit(X_TrSc, y_Train)
    print("Training Score: ", model.score(X_TrSc,y_Train))
    print()
    print("Test Score: ",model.score(X_TeSc,y_Test))
    print()
    print("Cross Validation Score: ", cross_val_score(model,X_TrSc,y_Train).mean())
    print()
    print('-'*60)
    print()
```

```python
from sklearn.metrics import r2_score

#we should evaluate the GLM models (poisson,tweedie,gamma) with r^2 for good measure
```

```python
models2 = [
    Lars(eps=0.1234),
    PoissonRegressor(),
    TweedieRegressor(link='log', power=1.5),
    GammaRegressor()
]

#while we're doing these models again we may as well try to alter the Lars model for better results
```

```python
for model in models2:
    print(model)
    print()
    model.fit(X_TrSc, y_Train)
    train_preds = model.predict(X_TrSc)
    test_preds = model.predict(X_TeSc)
    print("Training Score: ", r2_score(y_Train,train_preds))
    print()
    print("Test Score: ",r2_score(y_Test,test_preds))
    print()
    print("Cross Validation Score: ", cross_val_score(model,X_TrSc,y_Train,scoring='r2').mean())
    print()
    print('-'*60)
    print()
```

```python
#for optimisation let's take the Lars and Tweedie regressors since they are the best performing models even when the 
#scoring metric is r^2 for the tweedie models (NB that Poisson and Gamma are special versions of the Tweedie so we only need
#to consider Tweedie for optimisation) 
```

```python
ensemble_models = [
    BaggingRegressor(DecisionTreeRegressor(max_depth=7)),
    RandomForestRegressor(max_depth=7),
    ExtraTreesRegressor(max_depth=7),
    AdaBoostRegressor(DecisionTreeRegressor(max_depth=7)),
    GradientBoostingRegressor()
]
```

```python
for model in ensemble_models:
    print(model)
    print()
    model.fit(X_TrSc, y_Train)
    print("Training Score: ", model.score(X_TrSc,y_Train))
    print()
    print("Test Score: ",model.score(X_TeSc,y_Test))
    print()
    print("Cross Validation Score: ", cross_val_score(model,X_TrSc,y_Train).mean())
    print()
    print('-'*60)
    print()
```

```python
#from the ensemble models the ones that will be considered for optimisation are going to be Bagging, random forest and 
#gradient boosting as they all produce comparatively strong test scores
```
